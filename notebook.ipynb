```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("milknew.csv")
```

```python
df.head()
```

```python
df['Grade']=df['Grade'].map({"high":2,"low":0,"medium":1})
```

```python
x=df.iloc[:,:-1]
x
```

```python
y = df['Grade']
y
```

```python
y
```

```python
y.value_counts()
```

```python
x
```

```python
from sklearn.tree import DecisionTreeClassifier
model= DecisionTreeClassifier()
```

```python
from sklearn.linear_model import LogisticRegression
model2 = LogisticRegression(max_iter=1000,solver="liblinear")
```

```python
from sklearn.neighbors import KNeighborsClassifier
model3 = KNeighborsClassifier(n_neighbors=7)
```

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
valid = StratifiedKFold(10)
```

```python
result = cross_val_score(model,x,y,cv= valid)
result
result.mean()
```

```python
result2 = cross_val_score(model2,x,y,cv = valid)
result2
result2.mean()
```

```python
result3 = cross_val_score(model3,x,y,cv = valid)
result3
result3.mean()
```

```python
from sklearn.model_selection import train_test_split as tts
x_train,x_test,y_train,y_test = tts(x,y,train_size=0.8,random_state=0)
```

```python
x_train.shape,y_train.shape
```

```python
model.fit(x_train,y_train)
```

```python
y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print("Training")
print("Accuracy : ",accuracy_score(y_train,y_train_pred)*100)
print(confusion_matrix(y_train,y_train_pred))
```
